```python
import pandas as pd
pd.set_option('display.max_rows', 500)

df = pd.read_excel('final_dataset_20-21.xlsx')

df.rename(columns=lambda x: x.replace(" (20/21)", ""), inplace=True)

df.head(), df.columns
```

```python
for col in df.columns:
    print(col)
```

```python
df['Contract Years Left'] = pd.to_numeric(df['Contract Years Left'], errors='coerce')
df['Value'] = df['Value'] / 1000000

for col in df.columns:
    print(col)  

```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder()

columns_to_encode = ['League', 'Position']
encoded_data = encoder.fit_transform(df[columns_to_encode])
encoded_df = pd.DataFrame(encoded_data.toarray(), columns=encoder.get_feature_names_out(columns_to_encode))

df_categorical = pd.concat([df.drop(columns_to_encode, axis=1), encoded_df], axis=1)

df_numeric = df_categorical.select_dtypes(include=['float64', 'int64'])

df_numeric = df_numeric.dropna()

df_numeric


```
