```python
import pandas as pd

pd.set_option('display.max_rows', 500)
df = pd.read_excel('final_dataset_20-21.xlsx')



df = df.dropna()
```

```python
import numpy as np

df.rename(columns=lambda x: x.replace(" (20/21)", ""), inplace=True)
df['Contract Years Left'] = pd.to_numeric(df['Contract Years Left'], errors='coerce')
df['Value'] = df['Value'] / 1000000
df = df.select_dtypes(include=[np.number])
```

```python

correlation_matrix = df.corr()

sorted_corr = correlation_matrix['Value'].sort_values(ascending=False).to_frame()

print(sorted_corr)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 30))
sns.heatmap(sorted_corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Posortowana korelacja zmiennych z wartoci rynkow pikarzy')
plt.show()
```
